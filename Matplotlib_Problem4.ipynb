```python
# importing library
import matplotlib.pyplot as plt

# Some data to display
x = [1, 2, 3]
y = [0, 1, 0]
z = [1, 0, 1]
#creating 2 subplots
fig, ax = plt.subplots(2)
#plot and show
ax[0].plot(x,y)
ax[1].plot(x,z)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,2,0.01)
y = 1 + np.sin(2 * np.pi * x)
fig,((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2)
ax1.plot(x,y,color = 'orange')
ax2.plot(x,y,color = 'red')
ax3.plot(x,y,color = 'blue')
ax4.plot(x,y,color = 'green')
ax5.plot(x,y,color = 'black')
ax6.plot(x,y,color = 'magenta')
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,ax = plt.subplots(2,2)
x = [1,2,3]
y = [4,5,6]
ax[0,0].plot(x,y)
ax[0,1].scatter(x,y)
ax[1,0].bar(x,y)
ax[1,1].hist([1,3,3,4,3,4,1,1,1,2,2,4])
# set titles for all the subplots
ax[0,0].set_title('line plot')
ax[0,1].set_title('scatter plot')
ax[1,0].set_title('bar plot')
ax[1,1].set_title('hitogram')
# Adjust layout to prevent overlap
plt.tight_layout()
#Display the graphs
plt.show()
```
