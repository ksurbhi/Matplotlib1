Line Plot


```python
# import pyplot from matplotlib
from matplotlib import pyplot as plt
X = [1,2,3,4,5]
y = [2,4,8,16,32]
# plot the line plot
plt.plot(X,y)
#show the plot
plt.show()
```

Bar Plot


```python
from matplotlib import pyplot as plt
X = [2,4,6,8,10]
y = [10,5,4,7,2]
# plot bar plot
plt.bar(X,y)
plt.show()
```

Histogram

```python
from matplotlib import pyplot as plt
y = [10,5,4,7,2,10,10]
# plot histogram plot (tells the frequency of occurrance of each element)
plt.hist(y) # all values in y accured once except 10, 10 occured 3 times
plt.show()
```

Scatter Plot

```python
from matplotlib import pyplot as plt
X = [2,4,6,8,10]
y = [10,5,4,7,2]
# plot bar plot
plt.scatter(X,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plotting the data points. The first list [1,2,3,4] represents the x-axis values,
# and the second list [1,4,9,16] represents the corresponding y-axis values.
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

# Setting the range for the x-axis and y-axis.
# [0, 6, 0, 20] means the x-axis will range from 0 to 6 and the y-axis will range from 0 to 20.
plt.axis([0, 6, 0, 20])

# Display the plot.
# plt.show is a method, so you need to call it using parentheses ().
plt.show()
```

Pai chart

```python
from matplotlib import pyplot as plt
labels = ['Python','Java','C++']
sizes = [35,35,30]
colors = ['yellow', 'green', 'coral']
# Plotting the pie chart
plt.pie(x = sizes,labels = labels,colors= colors,autopct='%1.1f%%', startangle= 90)
plt.title('Pie Chart Example')
plt.show()
```

Area Chart

```python
import matplotlib.pyplot as plt
X = [1,2,3,4]
y1 = [5,10,15,20]
y2 = [4,8,12,16]
#area chart
plt.fill_between(X,y1,y2,color = 'gold',label = 'area btw 1 & 2')
plt.plot(X,y1,'bo',label = 'line 1',linewidth = 2)
plt.plot(X,y2,'ro',label = 'line 2')
plt.title('Area plot')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
# Legend and Display
plt.legend()
plt.show()
```
