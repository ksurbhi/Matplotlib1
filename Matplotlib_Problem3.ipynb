```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
# ax = plt.axes([0.3,0.5,0.8,0.8]) # this is the one way to add axes
# we can also add axes to an existing figure using the add_axes() method.
# This method takes a list of four values: [left, bottom, width, height],
# where all values are in the range [0, 1] and represent the fraction of the figure's width and height.
ax = fig.add_axes([0.1,0.1,0.8,0.8])

ax.legend(labels = ('cosine', 'sine'),loc = 'upper left')
```

```python
import matplotlib.pyplot as plt
import numpy as np
X = np.linspace(-np.pi, np.pi,15) # generates 15 values evenly spaced between -pi to pi
cos = np.cos(X)
sin = np.sin(X)
fig = plt.figure()
ax = plt.axes([0.1,0.1,0.8,0.8])
ax.plot(X,cos,marker = 'o',linestyle = 'solid',color = 'red')
ax.plot(X,sin,marker = 'o',linestyle = 'dashed',color = 'blue')
ax.legend(labels=('Cosine Function', 'Sine Function'),loc = 'upper left')
ax.set_title('Graph of Trignimetric functions Sine and Cosine')
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = plt.axes([0.1,0.1,0.8,0.8])
X = np.linspace(-np.pi, np.pi,15) # generates 15 values evenly spaced between -pi to pi
cos = np.cos(X)
sin = np.sin(X)
ax.plot(X,cos,'ro-')
ax.plot(X,sin,'bs:')
ax.legend(labels=('Cosine Function', 'Sine Function'),loc = 'upper left')
ax.set_title('Graph of Trignimetric functions Sine and Cosine')
plt.show
```
