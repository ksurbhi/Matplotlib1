```python
import matplotlib.pyplot as plt
years = [1972,1982,1992,2002,2012]
e_India = [100.6, 158.61,306.8,400.5,800.10]
e_Bangladesh = [20.5,30.6,60.5,100.4,200.5]
plt.plot(years, e_India,label = 'India',color = 'orange',marker = 'o',linewidth = 2,linestyle = 'solid')
plt.plot(years,e_Bangladesh, label = 'Bangladesh', color = 'Green',linewidth = 2,linestyle = 'dashed',marker = 'o')
plt.title('Electricity consumption per capita of India vs Bangladesh')
plt.xlabel('Electricity consumption in kWh')
plt.ylabel('Years')
plt.legend()
plt.show()
```
